### Cleaning Data

Melakukan pembersihan data berdasarkan analisis penilaian data

Menghapus tabel yang tidak diperlukan

```python
del hour_df
```

Mengubah nama beberapa kolom

```python
#mengubah nama judul kolom agar lebih mudah dimengerti
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weathercond',
    'cnt': 'count'
}, inplace=True)
```

```python
day_df.head()
```

Mengubah tipe data dteday

```python
# mengubah tipe data dteday menjadi tipe data date
day_df['dateday'] = pd.to_datetime(day_df.dateday)
```

```python
day_df.info()
```

Mengubah nilai dari variabel season, weekday, month, dan weathercond

```python
#mengubah angka menjadi nama hari
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
```

```python
day_df.head()
```

```python
#mengubah angka menjadi nama bulan
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
```

```python
day_df.head()
```

```python
#mengubah angka menjadi jenis musim
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
```

```python
day_df.head()
```

```python
day_df['weathercond'] = day_df['weathercond'].map({
    1: 'Clear/Few clouds',
    2: 'Mist/Cloudy',
    3: 'Light snow/Rain',
    4: 'Heavy rain/Thunderstorm'
})
```

```python
day_df.head()
```

Mengubah nilai dari variabel yr

```python
day_df['year'] = day_df['dateday'].dt.year
```

```python
day_df.head()
```

Simpan data yang telah dibersihkan ke dalam folder dashboard

```python
day_df.to_csv('dashboard/clean_bike_share_data.csv', index=False)
```

## Exploratory Data Analysis (EDA)

###**Eksplorasi Data day_df**

1. Mengelompokkan peminjam sepeda (casual maupun terdaftar) berdasarkan bulan peminjaman

```python
day_df.groupby(by='month').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

Berdasarkan tabel hasil diatas dapat disimpulkan bahwa bulan Juni memiliki rata-rata dan jumlah peminjam terbesar. Sedangkan bulan Januari memiliki rata-rata dan jumlah peminjam terkecil.

2. Mengelompokkan peminjam sepeda (casual maupun terdaftar) berdasarkan keadaan cuaca

```python
day_df.groupby(by='weathercond').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

Berdasarkan tabel hasil diatas dapat disimpulkan peminjam cenderung meminjam ketika cuaca sedang cerah atau berawan. Sedangkan sangat sedikit saat salju turun atau hujan, dan tidak ada ketika cuaca sedang sangat buruk.

3. Mengelompokkan peminjam sepeda (casual maupun terdaftar) berdasarkan hari libur.

```python
day_df.groupby(by='holiday').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

Berdasarkan tabel hasil diatas dapat disimpulkan peminjam lebih kerap meminjam sepeda saat hari biasa, bukan hari libur.

4. Membandingkan peminjam sepeda pada saat hari kerja (weekday) dan akhir pekan (weekend).

```python
day_df.groupby(by='weekday').agg({
    'count':['max','min','mean']
})
```

Berdasarkan tabel hasil diatas urutan rata-rata jumlah peminjam sepeda dari yang terbesar ke yang terkecil adalah Jum'at(Fri), Kamis(Thu), Sabtu(Sat), Rabu(Wed), Selasa(Tue), Senin(Mon), dan Minggu(Sun).

5. Mengelompokkan peminjam sepeda (casual maupun terdaftar) berdasarkan hari kerja.

```python
day_df.groupby(by='workingday').agg({
    'count': ['max', 'min', 'mean']
})
```

Berdasarkan tabel hasil diatas dapat disimpulkan bahwa jumlah peminjaman sepeda terbesar terdapat di hari kerja atau working day dibandingkan dengan hari non working day.

6. Mengelompokkan peminjam sepeda (casual maupun terdaftar) berdasarkan musim.

```python
day_df.groupby(by='season').agg({
    'casual': 'mean',
    'registered': 'mean',
    'count': ['max', 'min', 'mean']
})
```

Berdasarkan tabel hasil diatas, dari rata-rata nya dapat disimpulkan bahwa baik peminjam casual maupun terdaftar suka meminjam pada musim gugur (Fall). Sedangkan musim yang paling sedikit peminjaman yaitu pada musim semi (Spring).
